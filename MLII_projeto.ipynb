## Preencher os valores dos nomes pela coluna "Costumer ID"

```python
# Para usar o metodo "ffill (preenche com os dados anteriores)" os dados precisam estar classificados e os valores NA´s precisam estar necessariamente no final, para um grupo não começar com NA

df_credito = df_credito.sort_values(by=['Customer_ID', 'Name'], na_position='last')


df_credito['Name'] = df_credito.groupby('Customer_ID')['Name'].transform(lambda x: x.fillna(method='ffill'))

```

```python
df_credito['Name'].isnull().sum()
```

```python
# Retirar os "and" e tirar o espaço após a virgula
# Aplica o get_dummies com separador "," 
# Concatena os dataframes e exclui a coluna original

df_credito['Type_of_Loan'] = df_credito['Type_of_Loan'].str.replace(' and ','')
df_credito['Type_of_Loan'] = df_credito['Type_of_Loan'].str.replace(', ',',')


df_cat = df_credito['Type_of_Loan'].str.get_dummies(sep=',')


df_credito = pd.concat([df_credito, df_cat], axis=1)
df_credito = df_credito.drop('Type_of_Loan', axis=1)

# Exibir o resultado
df_credito
```
